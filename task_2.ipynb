```python
def solution(pegs):
    consts = []
    isEven = len(pegs) % 2 == 0 # is it even

    # calulate c_i to c_n, the constants
    for idx in range(len(pegs)):
        if idx != len(pegs) - 1:
            consts.append(pegs[idx + 1] - pegs[idx])

    #  calculate r_n, the last radius, subtract the even constants. c_1 - c_2 + c_3 ...
    r_n = sum(consts[::2]) - sum(consts[1::2])
    a = r_n * 2

    if isEven:
        # if first radius is divisible by 3
        if (a / 3 < 1):
            return [-1, -1]
        elif (a % 3 == 0):
            r_1 = [a / 3, 1]
        else:
            r_1 = [a, 3]
    else:
        r_1 = a
        if (r_1 < 1):
            return [-1, -1]
        r_1 = [r_1, 1]
    
    r_i = r_1[0] / r_1[1]
    for const in consts:
        # check radius is correct
        if (r_i < 1):
            return [-1, -1]
        # calculate the next radius
        r_i = const - r_i

    return r_1
```
